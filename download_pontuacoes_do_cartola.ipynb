```python
import requests
import pandas as pd
import json
from google.colab import drive
```

```python
#https://api.cartola.globo.com/partidas/1
#https://api.cartola.globo.com/atletas/mercado
cartola = requests.get('https://api.cartola.globo.com/atletas/mercado')
```

```python
dados = json.loads(cartola.text)
```

```python
dados = pd.Series(dados).to_frame('dados')
```

```python
dados = dados.explode(column='dados' , ignore_index=True)
```

```python
data = pd.json_normalize(data=dados['dados'])
```

```python
data.head(32)
```

```python
# Excluir as primeiras linhas que retornaram com NA.
data = data.drop(labels=range(0, 31), axis=0)
```

```python
data.head()
```

```python
# Salvar diretamente no meu google drive
drive.mount ('/drive')
data.to_excel('/drive/MyDrive/Cartola_2022/rodada_38.xlsx')
```
